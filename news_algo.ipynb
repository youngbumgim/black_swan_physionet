```python
def timetosecond(df):
    # 일자와 시분초 구분
    splited_time = []
    for i in df['rec_time']:
        time = i.split(',')         # 1day와 시간 구분
        time = list(reversed(time)) # 1day가 뒤로 가게 변경
        splited_time.append(time)
    # 초 단위로 치환
    second_list = []
    for each in splited_time:
        int_list = []
        
        if len(each) == 2:
            time_split = each[0].split(':')    #time_split = [' 0', '14', '00']
            for i in time_split:            # int로 변환
                element = int(i)
                int_list.append(element) # int_list = [ 0, 14, 00]
            int_list[0] = int_list[0]+24

        else:
            time_split = each[0].split(':')
            for i in time_split:        # int로 변환
                element = int(i)
                int_list.append(element)

        h_to_m = int_list[0]*60
        m = int_list[1] + h_to_m
        second = m*60 + int_list[2]
        second_list.append(second)
    # 초단위를 rec_time에 넣기
    rec_second = pd.DataFrame(second_list, columns={'rec_second'})
    df['rec_time'] = rec_second
    # 피드순, 시간 순 정렬
    df.sort_values(by=['pid','rec_time'])
        
    return df
```

```python
df.loc[df['RR']<=8, 'rr_score'] = 3
df.loc[(df['RR']>=9) & (df['RR']<=11), 'rr_score'] = 1
df.loc[(df['RR']>=12) & (df['RR']<=20), 'rr_score'] = 0
df.loc[(df['RR']>=21) & (df['RR']<=24), 'rr_score'] = 2
df.loc[df['RR']>=25, 'rr_score'] = 3

df.loc[df['SPO2']<=91, 'os_score'] = 3
df.loc[(df['SPO2']>=92) & (df['SPO2']<=93), 'os_score'] = 2
df.loc[(df['SPO2']>=94) & (df['SPO2']<=95), 'os_score'] = 1
df.loc[df['SPO2']>=96, 'os_score'] = 0


df.loc[df['BT']<=35, 'bt_score'] = 3
df.loc[(df['BT']>=35.1)&(df['BT']<=36.0) , 'bt_score'] = 1
df.loc[(df['BT']>=36.1)&(df['BT']<=38.0) , 'bt_score'] = 0
df.loc[(df['BT']>=38.1)&(df['BT']<=39.0) , 'bt_score'] = 1
df.loc[df['BT']>=39.1, 'bt_score'] = 2

df.loc[df['SBP']<=90, 'SBP_score'] = 3
df.loc[(df['SBP']>=91) & (df['SBP']<=100), 'SBP_score'] = 2
df.loc[(df['SBP']>=101) & (df['SBP']<=110), 'SBP_score'] = 1
df.loc[(df['SBP']>=111) & (df['SBP']<=219), 'SBP_score'] = 0
df.loc[df['SBP']>=220, 'SBP_score'] = 3

df.loc[df['HR']<=40, 'HR_score'] = 3
df.loc[(df['HR']>=41) & (df['HR']<=50), 'HR_score'] = 1
df.loc[(df['HR']>=51) & (df['HR']<=90), 'HR_score'] = 0
df.loc[(df['HR']>=91) & (df['HR']<=110), 'HR_score'] = 1
df.loc[(df['HR']>=111) & (df['HR']<=130), 'HR_score'] = 2
df.loc[(df['HR']>=131), 'HR_score'] = 3

df.loc[df['GCS']>=13, 'consciousness_score'] = 0
df.loc[(df['GCS']<=12), 'consciousness_score'] = 3
```

```python
df['NEWS_score'] = df['rr_score'] + df['os_score'] + df['bt_score'] + df['SBP_score'] + df['HR_score'] + df['consciousness_score']
```

```python
import numpy as np
news_max_score_list = []
news_min_score_list = []
news_mean_score_list = []
news_median_score_list = []
for i in df['pid'].unique():
    news_max_score_list.append(np.max(df[df['pid']==i]['NEWS_score']))
    news_min_score_list.append(np.min(df[df['pid']==i]['NEWS_score']))
    news_mean_score_list.append(np.mean(df[df['pid']==i]['NEWS_score']))
    news_median_score_list.append(np.median(df[df['pid']==i]['NEWS_score']))
```

```python
test_y_list = []
for pid in test_df['pid'].unique():
    test_y_list.append(test_df[test_df['pid']== pid]['target'].values[0])
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc
from sklearn import metrics
```

```python
max_auroc_score = metrics.roc_auc_score(test_y_list, news_max_score_list)
min_auroc_score = metrics.roc_auc_score(test_y_list, news_min_score_list)
mean_auroc_score = metrics.roc_auc_score(test_y_list, news_mean_score_list)
median_auroc_score = metrics.roc_auc_score(test_y_list, news_median_score_list)

print(max_auroc_score)
print(min_auroc_score)
print(mean_auroc_score)
print(median_auroc_score)
```
