```python
import pandas as pd

a = pd.read_csv('C:/project/physionet2012/physionet2012_train.csv', encoding='cp949')
```

```python
df = pd.DataFrame()
for col in a.columns:
    b = [[col, round(100 * (a[col].isnull().sum() / a[col].shape[0]),2)]]
    b = pd.DataFrame(b)
    df = pd.concat([df,b])
    # print(b)
```

```python
df.columns = ['var', 'percent']
```

```python
sorted_df = df.sort_values('percent')
```

```python
#90%이상인 컬럼 추출
a = sorted_df[sorted_df['percent']>99]
a['var'].to_list()
```
