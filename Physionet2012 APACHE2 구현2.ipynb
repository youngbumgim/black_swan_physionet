```python
import pandas as pd
```

```python
df = pd.read_csv("df_survival__.csv")
```

```python
df
```

```python
df5 = pd.read_csv("df5.csv")
```

```python
df5
```

```python
df = pd.concat([df5,df], axis = 1)
```

```python
df
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score
```

```python
#mean이용

df['AP_patient_mean'] = df.groupby('pid')['AP_sum'].transform(np.mean)
df
```

```python
df['survival__'].values
```

```python
drop_df = df5.drop_duplicates('pid')
auc_score = roc_auc_score(df['survival__'], df['AP_patient_mean'])
auc_score
```

```python
#max이용

df['AP_patient_max'] = df.groupby('pid')['AP_sum'].transform(np.max)
drop_df = df5.drop_duplicates('pid')
auc_score = roc_auc_score(df['survival__'], df['AP_patient_max'])
auc_score
```

```python
#minimum이용

df['AP_patient_minimum'] = df.groupby('pid')['AP_sum'].transform(np.min)
drop_df = df5.drop_duplicates('pid')
auc_score = roc_auc_score(df['survival__'], df['AP_patient_minimum'])
auc_score
```

```python
#median 이용

df['AP_patient_median'] = df.groupby('pid')['AP_sum'].transform(np.median)
drop_df = df5.drop_duplicates('pid')
auc_score_median = roc_auc_score(df['survival__'], df['AP_patient_median'])
auc_score_median
```

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt

RocCurveDisplay.from_predictions(df['survival__'], df['AP_sum'], name='APACHE2')
plt.show()
```
